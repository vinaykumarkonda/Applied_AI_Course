```
# Plot missing values for features

def plot_missing_values(data):
  total_records = data.shape[0]
  feature_names = data.columns.values
  missing = []
  for feature_name in feature_names:
    missing_percent = round((total_records-data[feature_name].count())/total_records * 100,2)
    missing.append(missing_percent)
  plt.bar(feature_names,missing)

plot_missing_values(tweets_test_data)
```

```
#stacked bar plots matplotlib: https://matplotlib.org/gallery/lines_bars_and_markers/bar_stacked.html
def stack_plot(data, xtick, col2='target', col3='total'):
    ind = np.arange(data.shape[0])
    
    plt.figure(figsize=(20,5))
    p1 = plt.bar(ind, data[col3].values)
    p2 = plt.bar(ind, data[col2].values)

    plt.ylabel('Tweets')
    plt.title('% of real disaster tweets keyword wise')
    plt.xticks(ind, list(data[xtick].values), rotation=45)
    plt.legend((p1[0], p2[0]), ('tweets', 'real disaster'))
    plt.show()
```

```
def univariate_barplots(data, col1, col2='target', top=False):
    # Count number of zeros in dataframe python: https://stackoverflow.com/a/51540521/4084039
    temp = pd.DataFrame(data.groupby(col1)[col2].agg(lambda x: x.eq(1).sum())).reset_index()

    # Pandas dataframe grouby count: https://stackoverflow.com/a/19385591/4084039
    temp['total'] = pd.DataFrame(data.groupby(col1)[col2].agg({'total':'count'})).reset_index()['total']
    temp['Avg'] = pd.DataFrame(data.groupby(col1)[col2].agg({'Avg':'mean'})).reset_index()['Avg']
    
    temp.sort_values(by=['total'],inplace=True, ascending=False)
    
    if top:
        temp = temp[0:top]
    
    stack_plot(temp, xtick=col1, col2=col2, col3='total')
    print(temp.head(5))
    print("="*50)
    print(temp.tail(5))
```

# Text Preprocessing

```
STOP_WORDS = stopwords.words("english")


def preprocess(x):
    x = str(x).lower()
    x = x.replace(",000,000", "m").replace(",000", "k").replace("", "'").replace("", "'")\
                           .replace("won't", "will not").replace("cannot", "can not").replace("can't", "can not")\
                           .replace("n't", " not").replace("what's", "what is").replace("it's", "it is")\
                           .replace("'ve", " have").replace("i'm", "i am").replace("'re", " are")\
                           .replace("he's", "he is").replace("she's", "she is").replace("'s", " own")\
                           .replace("%", " percent ").replace("", " rupee ").replace("$", " dollar ")\
                           .replace("", " euro ").replace("'ll", " will")
    x = re.sub(r"([0-9]+)000000", r"\1m", x)
    x = re.sub(r"([0-9]+)000", r"\1k", x)
    
    
    porter = PorterStemmer()
    pattern = re.compile('\W')
    
    if type(x) == type(''):
        x = re.sub(pattern, ' ', x)
    
    
    if type(x) == type(''):
        x = porter.stem(x)
        example1 = BeautifulSoup(x)
        x = example1.get_text()
               
    
    return x
```

```
# Text pre-processing

preprocessed_train_tweets = tweets_train_data['text'].apply(preprocess)
train_target = tweets_train_data['target']

preprocessed_test_tweets = tweets_test_data['text'].apply(preprocess)
```

## Bag of Words

```
# We are considering only the words which appeared in at least 10 documents(rows or projects).
bow_vectorizer = CountVectorizer(min_df=10, ngram_range=(1,4))
bow_x_train = bow_vectorizer.fit_transform(X_train)

bow_x_test = bow_vectorizer.transform(X_test)

print("Shape of BOW X_train matrix after text pre-processing ",bow_x_train.shape)

print("Shape of BOW X_test matrix after text pre-processing ",bow_x_test.shape)

bow_test_tweets = bow_vectorizer.transform(preprocessed_test_tweets)

print("Shape of BOW Test matrix after text pre-processing ",bow_test_tweets.shape)
```

## TFIDF vectorizer

```
tfidf_vectorizer = TfidfVectorizer(min_df=10, ngram_range=(1,4))
tfidf_x_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_x_test = tfidf_vectorizer.transform(X_test)

print("Shape of TFIDF X_train matrix after text pre-processing ",tfidf_x_train.shape)

print("Shape of TFIDF X_test matrix after text pre-processing ",tfidf_x_test.shape)

# we are converting a dictionary with word as a key, and the idf as a value
dictionary = dict(zip(tfidf_vectorizer.get_feature_names(), list(tfidf_vectorizer.idf_)))
tfidf_words = set(tfidf_vectorizer.get_feature_names())

tfidf_test_tweets = tfidf_vectorizer.transform(preprocessed_test_tweets)

print("Shape of TFIDF Train matrix after text pre-processing ",tfidf_test_tweets.shape)
```

```
def tune_hyperparameter(classifier, nlp_type, parameters, x_train, y_train):
    hyper_type = list(parameters.keys())[0]
    
    ## The hyper paramter tuning(find best alpha:smoothing parameter)
    
    clf = GridSearchCV(classifier, parameters, cv=3, return_train_score=True, scoring='roc_auc')
    clf.fit(x_train, y_train)

    results = pd.DataFrame.from_dict(clf.cv_results_).sort_values(['param_alpha'])

    train_auc = results['mean_train_score']
    train_auc_std = results['std_train_score']
    cv_auc = results['mean_test_score'] 
    cv_auc_std= results['std_test_score']
    hyperparameter =  results['param_'+hyper_type]

    plt.plot(hyperparameter, train_auc, label='Train AUC')
    plt.plot(hyperparameter, cv_auc, label='CV AUC')

    plt.scatter(hyperparameter, train_auc, label='Train AUC points')
    plt.scatter(hyperparameter, cv_auc, label='CV AUC points')

    plt.legend(loc=0)
    plt.xlabel(hyper_type+": hyperparameter")
    plt.ylabel("AUC")
    plt.xscale('log')
    plt.title("{} - Hyper parameter Vs AUC plot".format(nlp_type))

    # Best alpha hyperparameter with accuracy

    auc, best_hyperparemeter  = sorted(list(zip(cv_auc, hyperparameter)),reverse=True)[0]
    print("Optimal hyperparameter for '{}' is {} with accuracty {}".format(hyper_type, best_hyperparemeter, auc))
    
    return best_hyperparemeter
```

```
def roc_auc_metric(clf, nlp_type, x_train, y_train, x_test, y_test):
    
    clf.fit(x_train, y_train)

    y_train_pred = clf.predict_proba(x_train)[:,1]    
    y_test_pred = clf.predict_proba(x_test)[:,1]

    train_fpr, train_tpr, tr_thresholds = roc_curve(y_train, y_train_pred)
    test_fpr, test_tpr, te_thresholds = roc_curve(y_test, y_test_pred)

    train_auc = auc(train_fpr, train_tpr)
    test_auc = auc(test_fpr, test_tpr)
    
    print("Train AUC is {} and Test AUC is {}".format(train_auc, test_auc))

    plt.plot(train_fpr, train_tpr, label="train AUC ="+str(train_auc))
    plt.plot(test_fpr, test_tpr, label="test AUC ="+str(test_auc))
    plt.legend()
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.title("{} - ROC-AUC Curve".format(nlp_type))
    
    return train_auc, test_auc
```

# Multinomial NB

```
parameters = {'alpha':[0.00001,0.0001,0.001,0.01,0.1,1,10,100,1000,10000]}

# BOW

bow_hyperparemeter = tune_hyperparameter(MultinomialNB(), 'BOW', parameters, bow_x_train, y_train)
```

```
multi_nb = MultinomialNB(alpha=bow_hyperparemeter)

bow_train_auc, bow_test_auc = roc_auc_metric(multi_nb, "BOW", bow_x_train, y_train, bow_x_test, y_test )
```

```
# TFIDF

tfidf_hyperparemeter = tune_hyperparameter(MultinomialNB(), 'TFIDF', parameters, tfidf_x_train, y_train)
```

```
multi_nb = MultinomialNB(alpha=tfidf_hyperparemeter)

tfidf_train_auc, tfidf_test_auc = roc_auc_metric(multi_nb, "TFIDF", tfidf_x_train, y_train, tfidf_x_test, y_test )
```

```
t = PrettyTable(['Vectorizer', 'Model', 'Hyper parameter', 'Train AUC', 'Test AUC'],)
t.add_row(['BOW', 'MultiNB', bow_hyperparemeter, round(bow_train_auc,2), round(bow_test_auc,2)])
t.add_row(['TFIDF', 'MultiNB', tfidf_hyperparemeter, round(tfidf_train_auc,2), round(tfidf_test_auc,2)])
print(t)
```
